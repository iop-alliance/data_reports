```python
import pandas as pd
import concurrent.futures
import ast
import folium.plugins
```

```python
labs = pd.read_excel('fablabs.xlsx')
machines = pd.read_excel('machines.xlsx')

machines['labs'] = machines['labs'].apply(lambda x: ast.literal_eval(x))
labs['capabilities'] = labs['capabilities'].apply(lambda x: ast.literal_eval(x))

machines['name'] = machines['name'].apply(lambda x: x.lower())

# labs['name'] = labs['name'].apply(lambda x: x.lower().replace(' ', '').replace('fab', '').replace('lab', ''))

# machines['labs'] = machines['labs'].apply(lambda x: [lab.lower().replace(' ', '').replace('fab', '').replace('lab', '').replace('fablab', '') for lab in x])
```

```python
def filter_txt(txt):
    return txt.lower().replace(' ', '').replace('fab', '').replace('lab', '').replace('"', '').replace( "'", "")
```

```python
labs['name'] = labs['name'].apply(lambda name: filter_txt(name))
machines['labs'] = machines['labs'].apply(lambda x: [filter_txt(lab) for lab in x])
```

```python
import folium

# Create map centered on a specific location
map_center = [37.7749, -122.4194]
m = folium.Map(location=map_center, zoom_start=12)

# Add markers for machines with different capabilities
for _, lab in labs.iterrows():
    if lab['latitude']:
        if '3D printing' in lab['capabilities']:
            icon = folium.Icon(color='green', icon='cube', prefix='fa')
        elif 'Laser cutting' in lab['capabilities']:
            icon = folium.Icon(color='red', icon='cut', prefix='fa')
        elif 'CNC milling' in lab['capabilities']:
            icon = folium.Icon(color='blue', icon='gears', prefix='fa')
        elif 'Electronics' in lab['capabilities']:
            icon = folium.Icon(color='orange', icon='plug', prefix='fa')
        else:
            icon = folium.Icon(icon_size=(35, 45), color='beige', icon='location', prefix='fa')

        popup_html = f"""
        <b>{lab['name']}</b><br>
        Capabilities: {lab['capabilities']}<br>
        Location: {lab['city']}<br>
        Contact: {lab['phone']}, {lab['email']}<br>
        """
        popup = folium.Popup(popup_html, max_width=300)
        folium.Marker(
            location=[lab['latitude'], lab['longitude']],
            icon=icon,
            popup=popup
        ).add_to(m)

m
```
