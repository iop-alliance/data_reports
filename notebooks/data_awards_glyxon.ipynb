```python
import nltk
nltk.download('stopwords')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer

# preprocess text data
cnc_data['name'] = cnc_data['name'].str.lower().str.replace('[^\w\s]','').str.strip()
stop_words = set(stopwords.words('english'))
cnc_data['name'] = cnc_data['name'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
stemmer = SnowballStemmer("english")
cnc_data['name'] = cnc_data['name'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))

# vectorize text data using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(cnc_data['name'])

# cluster common words using k-means
kmeans = KMeans(n_clusters=10, random_state=0)
kmeans.fit(X)

# add cluster labels to dataframe
cnc_data['cluster'] = kmeans.labels_
```

```python
cnc_data['cluster']
```

```python
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"Cluster {i}:")
    print([vectorizer.get_feature_names()[ind] for ind in center.argsort()[-1:]])
```

```python
import seaborn as sns

cluster_counts = cnc_data.drop(columns='geometry').groupby('cluster').size().reset_index(name='count')

# Sort by frequency
cluster_counts = cluster_counts.sort_values('count', ascending=False)

# Plot frequency of clusters
sns.set_style('darkgrid')
sns.barplot(x='cluster', y='count', data=cluster_counts)
```

```python
sns.clustermap(cluster_counts, method='single', metric='euclidean', linewidth=0.5, figsize=(10, 10))
```

```python
cnc_data['country'] = ""
cnc_data['city'] = ""

for index, row in cnc_data.iterrows():
    location = geolocator.reverse((row['geometry'].y, row['geometry'].x), exactly_one=True)
    if location is not None:
        if 'address' in location.raw:
            if 'country' in location.raw['address']:
                cnc_data.at[index, 'country'] = location.raw['address']['country']
            if 'city' in location.raw['address']:
                cnc_data.at[index, 'city'] = location.raw['address']['city']
```
